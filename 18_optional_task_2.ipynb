Histogram

```python
plt.figure(figsize=(10,8))

# Age Histogram
plt.subplot(2,2,1)
plt.hist(df["age"], edgecolor="black")
plt.title("Age")

# Daily Sleep Hours Histogram
plt.subplot(2,2,2)
plt.hist(df["daily_sleep_hours"], edgecolor="black")
plt.title("Daily Sleep Hours")

# Daily Screen Time Hours Histogram
plt.subplot(2,2,3)
plt.hist(df["daily_screen_time_hours"], edgecolor="black")
plt.title("Daily Screen Time Hours")

# CGPA Histogram
plt.subplot(2,2,4)
plt.hist(df["cgpa"], edgecolor="black")
plt.title("CGPA")

plt.tight_layout()
plt.show()
```

Scatter plot

```python
# Age vs CGPA Scatter Plot
plt.scatter(df["age"], df["cgpa"])

plt.title("Age vs CGPA")
plt.xlabel("Age")
plt.ylabel("CGPA")

plt.show()
```

Box plot

```python
# Age Box Plot
plt.boxplot(df["age"])

plt.title("Age Box Plot")
plt.ylabel("Age")

plt.show()
```

Countplot

```python
plt.figure(figsize=(16,8))

# Gender Count Plot
plt.subplot(2,2,1)
sns.countplot(x="gender", data=df)
plt.title("Gender Count Plot")
plt.xticks(rotation=0, fontsize=8)

# Year of Study Count Plot
plt.subplot(2,2,2)
sns.countplot(x="year_of_study", data=df)
plt.title("Year of Study Count Plot")
plt.xticks(rotation=0, fontsize=8)

# Major Category Count Plot
plt.subplot(2,2,3)
sns.countplot(x="major_category", data=df)
plt.title("Major Category Count Plot")
plt.xticks(rotation=0, fontsize=8)

# Semester Type Count Plot
plt.subplot(2,2,4)
sns.countplot(x="semester_type", data=df)
plt.title("Semester Type Count Plot")
plt.xticks(rotation=0, fontsize=8)

plt.tight_layout()
plt.show()
```

heatmap

```python
# Correlation Heatmap (4 × 4)

plt.figure(figsize=(8,6))

corr = df[["age",
           "cgpa",
           "daily_sleep_hours",
           "daily_screen_time_hours"]].corr()

sns.heatmap(corr,
            annot=True,
            fmt=".2f",
            cbar=True)

plt.title("Correlation Heatmap")
plt.show()
```

Piechart

```python
# Gender Pie Chart
df["gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
# Before One-Hot Encoding
print("===== Before One-Hot Encoding =====")
print(df.head())

# One-Hot Encoding
df = pd.get_dummies(df, columns=[
    "gender",
    "year_of_study",
    "major_category",
    "has_part_time_job",
    "financial_stress",
    "semester_type"
])

# After One-Hot Encoding
print("\n===== After One-Hot Encoding =====")
print(df.head())
```

```python
# Before Standard Scaling
print("===== Before Standard Scaling =====")
print(df.head())

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[["age", "cgpa", "daily_sleep_hours", "daily_screen_time_hours"]] = scaler.fit_transform(
    df[["age", "cgpa", "daily_sleep_hours", "daily_screen_time_hours"]]
)

# After Standard Scaling
print("\n===== After Standard Scaling =====")
print(df.head())
```
