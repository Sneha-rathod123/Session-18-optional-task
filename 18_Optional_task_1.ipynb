Age histogram

```python
plt.hist(df["age"], edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

expenses histogram

```python
plt.hist(df["expenses"], edgecolor="black")
plt.title("Expenses Distribution")
plt.xlabel("Expenses")
plt.ylabel("Count")
plt.show()
```

Scatterplot

```python
plt.scatter(df["age"], df["expenses"])
plt.title("Age vs Expenses")
plt.xlabel("Age")
plt.ylabel("Expenses")
plt.show()
```

Boxplot

```python
plt.boxplot(df["expenses"])
plt.title("Box Plot of Expenses")
plt.ylabel("Expenses")
plt.show()
```

countplot

```python
sns.countplot(x="sex", data=df)
plt.title("Count of Sex")
plt.show()

sns.countplot(x="smoker", data=df)
plt.title("Count of Smoker")
plt.show()

sns.countplot(x="region", data=df)
plt.title("Count of Region")
plt.show()
```

Correlation Heatmap

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

Peichart

```python
plt.subplot(1, 2, 1)
df["smoker"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Smoker Distribution")
plt.ylabel("")

plt.subplot(1, 2, 2)
df["sex"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Sex Distribution")
plt.ylabel("")

plt.show()
```

Encoding

```python
print("Before One-Hot Encoding")
print(df.head())

df = pd.get_dummies(df, drop_first=True)

print("\nAfter One-Hot Encoding")
print(df.head())
```

Standard scalling

```python
from sklearn.preprocessing import StandardScaler

print("Before Feature Scaling")
print(df.head())

scaler = StandardScaler()

df[["age", "bmi", "children", "expenses"]] = scaler.fit_transform(
    df[["age", "bmi", "children", "expenses"]]
)

print("\nAfter Feature Scaling")
print(df.head())
```
