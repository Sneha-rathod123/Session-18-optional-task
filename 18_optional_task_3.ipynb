Histogram

```python
plt.hist(df["Percent AI Service Focus"], edgecolor="black")
plt.title("AI Service Focus")
plt.xlabel("AI Service Focus (%)")
plt.ylabel("Frequency")

plt.xticks([])   # x-axis labels hide ho jayenge

plt.show()
```

Satterplot

```python
plt.scatter(df["Average Hourly Rate"], df["Percent AI Service Focus"])

plt.title("Hourly Rate vs AI Service Focus")
plt.xlabel("Average Hourly Rate")
plt.ylabel("Percent AI Service Focus")

plt.xticks(rotation=90, fontsize=7)
plt.yticks(fontsize=7)
plt.tight_layout()

plt.show()
```

Countplot

```python
sns.countplot(x="Number of Employees", data=df)

plt.title("Number of Employees")
plt.xticks(rotation=45)

plt.show()
```

Heatmap

```python
import seaborn as sns

# Convert all object columns to numeric codes
df_heatmap = df.copy()

for col in df_heatmap.columns:
    df_heatmap[col] = df_heatmap[col].astype("category").cat.codes

# Correlation Heatmap
sns.heatmap(df_heatmap.corr(), annot=True, cmap="Blues")

plt.title("Correlation Heatmap")
plt.show()
```

Piechart

```python
df["Location"].value_counts().head(10).plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Top 10 Company Locations")
plt.ylabel("")
plt.show()
```

```python
# One-Hot Encoding
df = pd.get_dummies(
    df,
    columns=["Location", "Number of Employees"],
    drop_first=True
)

# Output
print(df.head())
```

```python
from sklearn.preprocessing import StandardScaler

# One Hot Encoding (convert categorical data into numbers)
df_encoded = pd.get_dummies(df)

# Standard Scaling
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)

# Convert back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=df_encoded.columns)

# Output
print(scaled_df.head())
```
